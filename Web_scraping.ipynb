## World Bank API - Problem Statement
**This project focuses on understanding what factors influence GDP growth across countries using World Bank data.**
**The goal of this project is to analyze global economic indicators such as GDP, population, inflation, and unemployment across countries and years to identify trends and build predictive insights for economic growth.**

```python
 import requests  ## to call API
 import pandas as pd  ## to structure data
```

```python
def fetch_indicator(indicator_code, column_name):  
    ## It is reusable function to dynamically fetch different economic indicators.
    
## dynamically constructed API URLs to fetch different indicators with maximum records per request
 url = f"https://api.worldbank.org/v2/country/all/indicator/{indicator_code}?format=json&per_page=2000"

 response = requests.get(url)
 ## print("Status Code:", response.status_code)
 ## print("Response Preview:", response.text[:500])
 data = response.json()

# Added validation to handle API failures and prevent runtime errors.
 if not data or len(data) < 2:
    return pd.DataFrame()

 records = data[1]  ## actual data

    # Loop through each country-year data
 rows = []
 for item in records:
     
    rows.append({
        "country": item["country"]["value"],
        "year": item["date"],
       column_name: item["value"]  # dynamic column: makes function reusable
    })

 df = pd.DataFrame(rows)  ## convert list to data frame
 return df
```

```python
df_test = fetch_indicator("NY.GDP.MKTP.CD", "gdp")
print(type(df_test))
print(df_test.head())
```

```python
## Calling the Function
## Extracted multiple macroeconomic indicators to perform multivariate analysis.
df_gdp = fetch_indicator("NY.GDP.MKTP.CD", "gdp")
df_pop = fetch_indicator("SP.POP.TOTL", "population")
df_inflation = fetch_indicator("FP.CPI.TOTL.ZG", "inflation")
df_unemp = fetch_indicator("SL.UEM.TOTL.ZS", "unemployment")
df_gdp_growth = fetch_indicator("NY.GDP.MKTP.KD.ZG", "gdp_growth")


## Merge Data
## Merged multiple datasets on country and year to create a unified dataset for analysis
df = df_gdp.merge(df_pop, on=["country", "year"], how="left").merge(df_inflation, on=["country", "year"], how="left").merge(df_unemp, on=["country", "year"], how="left").merge(df_gdp_growth, on=["country", "year"], how="left")

print(df.head())
print("Total records:", len(df))
```

```python
## Coverting Data fame to csv file
df.to_csv("world_bank_data.csv", index=False)
```
